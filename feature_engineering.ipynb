```python
import pandas as pd
import numpy as np
```

## Task 1

Reading Values

```python
df = pd.read_csv("data.csv")
df.head()
```

Checking For Null

```python
df.isna().sum().sum()
```

Imputing Values

```python
cols = list(df)

for col in cols:
    if col=="diagnosis":
        df[col].fillna(value = df[col].mode(),inplace=True)
    elif col!="id":
        df[col].fillna(value = df[col].mean(),inplace=True)
```

Checking For Null

```python
df.isna().sum().sum()
```

Saving Data

```python
df.to_csv('data_fe1.csv')
```
